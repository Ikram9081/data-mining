```python
# Daftar kolom dengan nilai 0 tidak valid
cols_with_invalid_zeros = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

# Ganti 0 dengan NaN
df[cols_with_invalid_zeros] = df[cols_with_invalid_zeros].replace(0, np.nan)

# Isi NaN dengan median dari masing-masing kolom
for col in cols_with_invalid_zeros:
    df[col] = df[col].fillna(df[col].median())

# Pastikan semua NaN sudah hilang
print(df.isna().sum())  # Harus semua 0
```

```python
# Pisah fitur dan label
X = df.drop('Outcome', axis=1)
y = df['Outcome']

# Standarisasi fitur
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Evaluasi
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
y_pred = knn.predict(X_test)

print("Akurasi:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Simpan akurasi untuk setiap nilai k
k_values = range(1, 21)
accuracies = []

for k in k_values:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    y_pred_k = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred_k)
    accuracies.append(acc)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracies, marker='o', linestyle='-', color='blue')
plt.title("Akurasi vs Nilai K")
plt.xlabel("Jumlah Tetangga (k)")
plt.ylabel("Akurasi")
plt.xticks(k_values)
plt.grid(True)
plt.show()

# Tampilkan k terbaik
best_k = k_values[accuracies.index(max(accuracies))]
best_acc = max(accuracies)
print(f"Nilai k terbaik: {best_k} dengan akurasi: {best_acc:.4f}")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans


# 1. Load data
df = pd.read_csv("diabetes.csv")

# 2. Pilih fitur untuk clustering (tidak pakai kolom 'Outcome')
features = ['Glucose', 'BloodPressure', 'BMI', 'Age']
X = df[features].copy()

# 3. Ganti nilai 0 jadi NaN (karena tidak valid di medis)
X = X.replace(0, np.nan)
X.fillna(X.median(), inplace=True)

# 4. Standarisasi fitur
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 5. Clustering K-Means (misalnya 3 cluster)
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# 6. Visualisasi hasil clustering
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Glucose', y='BMI', hue='Cluster', palette='Set2')
plt.title("Hasil Clustering Pasien Diabetes")
plt.xlabel("Glukosa")
plt.ylabel("BMI")
plt.show()
```
